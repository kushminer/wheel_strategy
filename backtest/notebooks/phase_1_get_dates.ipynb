```python
# Step 1: Fetch historical close prices

tickers = ['SPY']
end_date = datetime(2025, 6, 1)
start_date = end_date - timedelta(days=365 * 10) # datetime(2025, 6, 1)
historic_close_data = fetch_historical_close_prices(tickers, start_date, end_date)

print(historic_close_data.head(3))
```

```python
# Step 2: Calculate technical indicators for multiple configurations

# 1. Define configurations for sma and bb
# 2. Calculate technical indicators for each configuration

technical_indicators_configs = {'config_ti1': {'window': 20, 'std_dev': 1.0}, 'config_ti2': {'window': 20, 'std_dev': 2.0},}
daily_tech_indicators_dict = calculate_technical_indicators(
    historic_close_data, 
    technical_filter_configs=technical_indicators_configs, 
    print_sample=False
)
```

```python
# Step 3: Filter daily tickers by technical indicators

# Define filter configurations

technical_indicator_filter_configs = {
    'config_tif1': lambda df: df['close'] <= df['bb_upper'],
    'config_tif2': lambda df: df['close'] <= df['bb_lower'],
    'config_tif3': lambda df: (df['close'].shift(-1) <= df['bb_lower'].shift(-1)) & (df['close'] >= df['bb_lower']),
}

filtered_daily_tickers_dict = filter_daily_tickers_by_technical(
    daily_tech_indicators_dict, 
    historic_close_data,
    filter_configs=technical_indicator_filter_configs
)
filtered_daily_tickers_dict['config_ti1']['config_tif1']
```

```python
# Step 4: Extract entry dates

entry_dates_dict = extract_entry_dates(filtered_daily_tickers_dict)
entry_dates_dict_str = extract_entry_dates(filtered_daily_tickers_dict, as_string=True)

# Display entry dates for each config
for tech_config_name, filter_configs_dict in entry_dates_dict_str.items():
    print(f"\n{tech_config_name}:")
    for filter_config_name, symbols_dict in filter_configs_dict.items():
        print(f"\n  {filter_config_name}:")
        for symbol, dates_list in symbols_dict.items():
            print(f"    {symbol}: {len(dates_list)} entry dates")
            print(f"      First 5 dates: {dates_list[:5]}")
```
